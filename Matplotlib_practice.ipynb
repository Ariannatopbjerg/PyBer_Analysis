```python
# What is in next cell:

# The data is in the form of arrays, or lists. 
# One array will contain the months of the year
# second will contain the total fares in US dollars for each month.
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Chaining the plot() function to plt
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Creating a line chart using Object-Oriented Interface
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()  # create the figure
ax.plot(x_axis, y_axis) # defining the axes
```

```python
# Let's break down what the code is doing:

# The plt.subplots() function returns a tuple that contains a figure and axes object(s).
# fig, ax = plt.subplots() unpacks the tuple into the variables fig and ax.
# The first variable, fig, references the figure.
# The second variable, ax, references the axes of the figure and allows us to annotate the figure.
# The ax.plot(x_axis, y_axis) plots the data from the x and y axes.
# The fig, ax = plt.subplots() line of code can also be written like this.

# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()

# Where fig references the figure, and ax references the axes of the figure and allows us 
#to annotate the figure.
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
  # Will print out a blank graph
# Then add...
ax.plot(x_axis, y_axis)
  # Have line within graph
```

```python
# Changing width and color of the line and markers -> declare parameters in the 'plt.plot()' 
# To declare parameters for line color, width, and marker, we use 'color=', 'width=', and 
# 'marker=', respectively, inside the 'plt.plot()' function.

# Create the plot.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

```python
# Scatter plot -> two ways: plt.plot() or plt.scatter()
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Add a lowercase "o" as a parameter inside the parentheses. 
# This switches the plot from a line chart to a scatter plot.
plt.plot(x_axis, y_axis, 'o')
```

```python
#Change the color of the markers to red.
plt.scatter(y_axis, x_axis, marker="o", c="red", label="Chicago")
#Add a legend for the city of Chicago.
plt.legend()
#Add a title and axes labels.
plt.title("PyBer Fare by Month")
plt.xlabel("Fare($)")
plt.ylabel("Date")
#Switch the axis so that the Fare($) data is on the x-axis.
#Add a limit for the x-axis data.
plt.xlim(0,50)
#Invert the y-axis so that January is at the top.
plt.gca().invert_yaxis()
```

```python
Create a Bubble Chart Using the MATLAB Method
```

```python
# s -> size parameter
plt.scatter(x_axis, y_axis, s=y_axis)
y_axis_larger = []
for data in y_axis:
  y_axis_larger.append(data*3)
```

```python
Create a Pie Chart Using the MATLAB Method
```

```python
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
# add the percentage of each wedge on the pie chart -> autopct parameter and provide the 
# format of one decimal place using .1f%.
plt.subplots(figsize=(8, 8))
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
```

```python
# Cloros are important 
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```
